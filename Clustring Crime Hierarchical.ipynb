```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
df= pd.read_csv('D:/Data Science/Assignment/07Clustring/crime_data.csv')
```

```python
df.isna().sum()
```

There is no null value so no need of imputation.

```python
x= df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
#normalize the data
sc_x = StandardScaler()
x=sc_x.fit_transform(x)
```

```python
#data visulize in dendogram
den=sch.dendrogram(sch.linkage(x,method='single'))
```

```python
hc= AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
#prediction
y_pr= hc.fit_predict(x)
```

```python
df['cluster']=hc.labels_
```
